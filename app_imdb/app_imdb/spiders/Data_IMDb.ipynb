```python
imdb = pd.read_csv("data_csv.csv")
imdb.head()
```

```python
# D'abord je trie  : 
# Je supprime tout les non digits du coût du film. (Pour la question bonus de fin) 
imdb['movie cost'] = imdb.loc[:,'movie cost'].replace(regex=r'\D+', value="")
imdb['movie cost']
```

```python
# Pour supprimer les doublons on peut utiliser imdb.drop_duplicates()
# Shape permet de voir le nombre de lignes et de colonnes. 
# Ici j'ai bien 250 films je n'ai pas besoin de la fonction

# imdb.drop_duplicates( inplace=True)

# Je trie la colonne durée 
imdb['duration'] = imdb.loc[:,'duration'].replace(regex=r",", value="").replace(regex=r" ", value="").replace(regex=r"m", value="")
imdb['duration']
```

```python

# Ma fonction qui change ma colonne heure en minute :

def to_minute(formatted_str):
    for i,j in formatted_str.items():
        if len(j) > 2  :
            cache = j.split("h")
        formatted_str[i] = (int(cache[0]) * 60 + int(cache[1])) 
    return formatted_str
to_minute(imdb['duration'])

# # Je transforme la colonne durée en int 

imdb['duration'] = pd.to_numeric(imdb['duration'])
```

```python
# Je sais récupéré des listes de chaque cellules 
actor_data =  imdb['actors']
list_actors = []
for i in actor_data:
    for j in i.split(","):
        list_actors.append(j)

category_data = imdb["genre"]
list_category = []
for i in category_data:
    for j in i.split(','):
        list_category.append(j)

language_data = imdb['Origin country']
list_language = []
for i in language_data:
    for j in i.split(','):
        list_language.append(j)
```

### 3 - Dans combien de films à joué Morgan Freeman ? et Tom Cruise ?  

```python
# La methodes plus adéquate ? 
a = imdb[imdb['actors'].str.contains("Morgan Freeman")]
b = imdb[imdb.loc[:,'actors'].str.contains("Tom Cruise")]
a
```

```python
# L'autre méthode : 
# fonctions pour compter les nombre de films par personnage

def actor_count(star, liste):
    count = 0 
    for i in liste:
        if star in str(i) :
            count +=1 
    return count 
```

```python
a = actor_count("Morgan Freeman", imdb['actors'] )    
b = actor_count("Tom Cruise", imdb['actors'] )  
print(f"Réponse 3 : Morgan Freeman a jouer dans {a} films et Tom Cruise a jouer dans {b} films. ")
```

```python
# boucle qui va remplir la colonne prix / minutes : 
imdb['movie cost by minute'] = 0 
for i,j in imdb['movie cost'].items():
    if j != 0  :
        imdb['movie cost by minute'][i] += float(imdb['movie cost'][i]) / float(imdb['duration'][i])
```
